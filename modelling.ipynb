# Simple example clustering model
An example of how to use this data to create a rudimentary playing style model.
 - Pick features that describe a playstyle
 - Reduced dimensionality to visualise using TSNE
 - Train a clustering algorithm (e.g. KMeans, DBScan)
 - Then if I was using bokeh, data points could be examined to see what the characteristics are of each match.

```python
model_cols = [
    "ratio_Pressure_vertical_zone_forward",
    "pass_long_ratio",
    "pass_success_ratio",
    "pass_long_ratio"]
```

```python
model_df = wsl_features[model_cols]
```

```python
model_df.values
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(model_df.values)
tsne = TSNE(n_components=2)
reduced_tsne = tsne.fit_transform(model_df.values)
```

```python
sns.scatterplot(x=reduced_tsne[:, 0], y=reduced_tsne[:, 1], hue="team", data=wsl_features)
```

```python
mask = wsl_features.team.isin(["Manchester City WFC", "Arsenal WFC", "Chelsea FCW"])
sns.scatterplot(x=reduced_tsne[:, 0], y=reduced_tsne[:, 1], hue=mask)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=2)
```

```python
pred = dbs.fit_predict(reduced_tsne)
```

```python
sns.scatterplot(x=reduced_tsne[:, 0], y=reduced_tsne[:, 1], hue=pred)
```
